```python
import pandas as pd
import numpy as np
from google_play_scraper import app,Sort,reviews_all
import plotly.express as px
```

```python
threads=reviews_all('com.instagram.barcelona',sleep_milliseconds=0,lang ='eng',country='US',sort=Sort.NEWEST)
```

```python
threads
```

#Creating a DataFrame

```python
df=pd.json_normalize(threads)
```

```python
df
```

```python
df['score'].mean()
```

```python
df['reviewCreatedVersion'].value_counts()
```

#Running Pre-trained Deep Learning Model




```python
from transformers import pipeline
sentiment_analysis= pipeline("sentiment-analysis",model="siebert/sentiment-roberta-large-english")
```

```python
print(sentiment_analysis("I love motogp so much"))
```

```python
df.dtypes
```

```python
df['content']=df['content'].astype('str')
```

```python
df1=df.head(1000)
```

```python
df1
```

```python
df1['result'] = None
```

```python
for index, row in df1.iterrows():
    df1.loc[index, 'result'] = sentiment_analysis(row['content'])
```

```python
df1=pd.DataFrame(df1)
```

```python
df1.head
```

```python
df1['sentiment']=df1['result'].apply(lambda x:(x[0]['label']))
df1['score']=df1['result'].apply(lambda x:(x[0]['score']))
```

```python
df1.head
```

```python
df1['score'].mean()
```

```python
df1['sentiment'].value_counts(normalize=True)
```

#Creating Visualization

```python
fig=px.histogram(df1,x='sentiment',color='sentiment',text_auto=True)
fig.show()
```
